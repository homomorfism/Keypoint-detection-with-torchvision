```python
from models.dataset import ChessDataset
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = ChessDataset(root_folder=Path("../data"), flag='train')
```

```python
n_pictures = 4
fig, axes = plt.subplots(ncols=4, nrows=1, figsize=(10 ,10))

for ii in range(n_pictures):
    image, label = dataset[ii + 30]

    image = np.transpose(image.numpy(), axes=(1, 2, 0))
    keypoints = label['keypoints'].squeeze()

    axes[ii].grid(False)
    axes[ii].set_xticks([])
    axes[ii].set_yticks([])
    axes[ii].imshow(image)
    axes[ii].scatter(keypoints[:, 0], keypoints[:, 1], color='red', s=100)



fig.savefig("train_data.png", bbox_inches='tight', pad=2)
```
