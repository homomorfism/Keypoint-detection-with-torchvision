```python
from models.dataset import ChessDataset
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = ChessDataset(root_folder=Path("../data"), flag='train')
```

```python
n_pictures = 4
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(10 ,10))

for ii in range(n_pictures):
    row = ii % 2
    column = ii // 2
    image, label = dataset[ii + 30]

    image = np.transpose(image.numpy(), axes=(1, 2, 0))
    keypoints = label['keypoints'].squeeze()

    axes[row, column].grid(False)
    axes[row, column].set_xticks([])
    axes[row, column].set_yticks([])
    axes[row, column].imshow(image)
    axes[row, column].scatter(keypoints[:, 0], keypoints[:, 1], color='red', s=100)



fig.savefig("train_data.png", bbox_inches='tight', pad=0)
```
